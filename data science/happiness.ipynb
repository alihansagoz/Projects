Question 5) Create a new column named as 'happiness_degree' which labels value of 'ladder'
*very happy if ladder < 30
*happy if 30 <= ladder < 60
*moderately happy if 60 <= ladder < 90
*unhappy if 90 <= ladder < 120
*very unhappy if ladder >= 120
Print the top 10 records updated dataframe (10 points)

```python
#codes
def label_race (row):
   if row['Ladder'] < 30 :
      return 'very happy'
   if 30 <= row['Ladder']  < 60:
      return 'happy'
   if 60 <= row['Ladder'] < 90 :
      return 'moderately happy'
   if 90 <= row['Ladder'] < 120 :
      return 'unhappy'
   if row['Ladder'] >= 120 :
      return 'very unhappy'

happiness_df['happiness_degree'] = happiness_df.apply (lambda row: label_race(row), axis=1)
happiness_df.head(10)
```

Question 6) Create heat map and comment which values are correlated with each other (15 points)

```python
#codes
corr = (happiness_df.iloc[:, 1:11:]).corr()
f, ax = plt.subplots(figsize=(12, 10))
mask = np.triu(np.ones_like(corr, dtype=bool))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, annot=True, mask = mask, cmap=cmap)
```
