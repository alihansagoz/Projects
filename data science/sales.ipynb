Question 1) Please read the "sales.csv" file as a DataFrame object named as "sales_df". Print the length and the first 5 tuples of the DataFrame. (5 points)

```python
#codes
sales_df =pd.read_csv('sales.csv')
sales_df.head()
```

Question2) Split the dataset into the independent (name it 'x') and dependent variables (name it 'y') which are predictors and response variable( 'Item_Outlet_Sales'). Print their shape info and top 10 records of them. (5 points)

```python
#codes2.1
x= sales_df.loc[:, sales_df.columns != 'Item_Outlet_Sales']
xshape = x.shape
print("Shape information x = " , xshape)
x.head(10)
```

```python
#codes2.2
y = sales_df[['Item_Outlet_Sales']]
yshape = y.shape
print("Shape information y = " , yshape)
y.head(10)
```

```python
# Importing MinMax Scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_norm = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
x_norm
```

Question 4) Split data into train and test sets. Use the split ratio as 80:20 and set the 'random_state' to an integer value. (10 points)

```python
# Importing Train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_norm,y,test_size=0.2,random_state = 1)
```

Question 5) Create an instance of kNN with a number of neighbors=3, then fit kNN regression model to train data. Finally, make prediction and calculate mean squared error (MSE). Print MSE value. (15 points)

```python
#importing KNN regressor and metric mse
from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.metrics import mean_squared_error as mse

knn = KNN(n_neighbors = 3)
knn.fit(x_train,y_train)
y_pred = knn.predict(x_test)
mse(y_test,y_pred)
```

Question 6) Create Elbow function which does the same operation in the previous question for each k instance in the range [1-40] by using Manhattan and Euclidean distance metrics separately in order to determine the optimal value of k. (25 points)

```python
#codes
error_rate = []
error_rate2 = []
# Will take some time
for i in range(1,40):
    knn = KNN(n_neighbors=i,metric='euclidean')
    knn.fit(x_train,y_train)
    pred_i = knn.predict(x_test)
    error_rate.append(mse(y_test,pred_i))
    
for i in range(1,40):
    knn = KNN(n_neighbors=i,metric='manhattan')
    knn.fit(x_train,y_train)
    pred_i = knn.predict(x_test)
    error_rate2.append(mse(y_test,pred_i))    
optimal_mse =999999999999; 
optimal_k = -1
distance_metric = ''
for i in range(len(error_rate)):
    if error_rate[i]<optimal_mse:
        distance_metric = 'euclidean'
        optimal_mse = error_rate[i]
        optimal_k = i+1
for i in range(len(error_rate2)):
    if error_rate[i]<optimal_mse:
        distance_metric = 'manhattan'
        optimal_mse = error_rate[i]
        optimal_k = i+1
print('Optimal Value of k for Elbow Function:')
print(optimal_k)        
```

```python
#codesplt.plot(range(1, 40),  error_rate, color ='g',label = 'Manhattan Distance')
plt.plot(range(1, 40), error_rate, color ='g',label = 'Manhattan Distance')
plt.plot(range(1, 40), error_rate2, color ='r',label = 'Euclidean Distance')
plt.xlabel("Value of K")
plt.ylabel("MSE Values")
plt.title('Mean Squared Error Values for kNN')
plt.legend()
plt.show() 
```
